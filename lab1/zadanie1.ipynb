```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
h = [np.power(10.,-k) for k in range(0,16)]
```

```python
def function(x) -> np.longdouble:
    return np.tan(x)
```

```python
def derrivative(h, x) -> np.longdouble:
    return (function(x+h) - function(x))/h
```

```python
actual_value =  np.longdouble(1+(np.tan(1)*np.tan(1)))
arrayOfValues = np.array([derrivative(h_value, 1) for h_value in h], dtype=np.longdouble)
```

```python
results =  np.array([np.abs(actual_value - arrayOfValues[i]) for i in range(0, 16)], dtype=np.longdouble)
```

```python
fig, plot = plt.subplots()
plot.loglog(h, results)

plot.set_xlabel("H Value")
plot.set_ylabel("Error Value")

plot.set_title("Wartość bezwględna błędu w zależonści od h")
```

```python
def machineEpsilon(func=float):
    machine_epsilon = func(1)
    while func(1)+func(machine_epsilon) != func(1):
        machine_epsilon_last = machine_epsilon
        machine_epsilon = func(machine_epsilon) / func(2)
    return machine_epsilon_last
```

```python
min_h = h[list(results).index(min(results))]
print(min_h)
```

```python
def derrivative2(h, x):
    return (function(x+h) - function(x-h))/(2*h)
```

```python
result2 = np.array([np.abs(derrivative2(h_value, 1) - actual_value) for h_value in h], dtype=np.longdouble)
```

```python
fig, plot = plt.subplots()
plot.loglog(h, result2)

plot.set_xlabel("h value")
plot.set_ylabel("Error Value")

plot.set_title("Wartość bezwględna błędu w zależonści od h")
```

```python
min_2h = h[list(result2).index(min(result2))]
print(min_2h)
```
